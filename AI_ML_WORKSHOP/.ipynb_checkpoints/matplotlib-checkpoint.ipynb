```python
import matplotlib.pyplot as plt
x = [3,4,5,8,5]
y = [5,6,7,5,1]
plt.plot(x,y)
plt.xlabel("rohit")    #TO give label to x axis
plt.ylabel("Arshu")    #To give label to y axis
plt.title("rohit-arshu graph")   #To give title to the graph
plt.show()    #TO show the graph area
```

```python
#sign graph
import numpy as np
x = np.arange(0,3*np.pi,0.1)
y = np.sin(x)
plt.plot(x,y)
plt.plot(x,np.cos(x))
plt.scatter(x,y,marker='1',c='red')
plt.legend(['Sine Graph','Cosicne'])
#TO remove the axis
plt.axis("off")
plt.show()
```
